```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import windows

# Define parameters
N = 128  # Number of points in the window

# Generate different windows
rect_window = windows.boxcar(N)
tri_window = windows.triang(N)
hamming_window = windows.hamming(N)
hanning_window = windows.hann(N)
kaiser_window = windows.kaiser(N, beta=8.6)  # Beta chosen for good side lobe suppression

windows_dict = {
    'Rectangular': rect_window,
    'Triangular': tri_window,
    'Hamming': hamming_window,
    'Hanning': hanning_window,
    'Kaiser': kaiser_window
}

# Plot each window separately
plt.figure(figsize=(12, 6))
for name, window in windows_dict.items():
    plt.plot(window, label=f'{name} Window')

plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.title('Window Functions')
plt.grid()
plt.legend()
plt.show()

# Plot FFT of each window in separate plots
for name, window in windows_dict.items():
    plt.figure(figsize=(10, 5))
    fft_output = np.fft.fft(window, 1024)  # Compute FFT with zero-padding
    fft_output = np.fft.fftshift(fft_output)  # Shift zero frequency to center
    magnitude_db = 20 * np.log10(np.abs(fft_output) / np.max(np.abs(fft_output)))
    freqs = np.linspace(-0.5, 0.5, len(fft_output))
    plt.plot(freqs, magnitude_db, label=f'{name} Window FFT')
    plt.xlabel('Normalized Frequency (cycles per sample)')
    plt.ylabel('Magnitude (dB)')
    plt.title(f'FFT of {name} Window')
    plt.grid()
    plt.legend()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import windows

# Define parameters
N = 128  # Number of points in the window
beta_values = [2, 5, 8.6, 14, 20]  # Different beta values for Kaiser window

# Plot Kaiser windows with different beta values
plt.figure(figsize=(12, 6))
for beta in beta_values:
    kaiser_window = windows.kaiser(N, beta)
    plt.plot(kaiser_window, label=f'Beta = {beta}')

plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.title('Kaiser Window with Different Beta Values')
plt.grid()
plt.legend()
plt.show()

# Plot FFT of Kaiser windows with different beta values in separate plots
for beta in beta_values:
    plt.figure(figsize=(10, 5))
    kaiser_window = windows.kaiser(N, beta)
    fft_output = np.fft.fft(kaiser_window, 1024)  # Compute FFT with zero-padding
    fft_output = np.fft.fftshift(fft_output)  # Shift zero frequency to center
    magnitude_db = 20 * np.log10(np.abs(fft_output) / np.max(np.abs(fft_output)))
    freqs = np.linspace(-0.5, 0.5, len(fft_output))
    plt.plot(freqs, magnitude_db, label=f'Beta = {beta}')
    plt.xlabel('Normalized Frequency (cycles per sample)')
    plt.ylabel('Magnitude (dB)')
    plt.title(f'FFT of Kaiser Window (Beta = {beta})')
    plt.grid()
    plt.legend()
    plt.show()
```
